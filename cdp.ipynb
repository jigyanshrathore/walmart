```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('C:\\Users\\HP\\Desktop\\walmart\\datasets\\cdp.csv' )

label_encoders = {}

le_program = LabelEncoder()
df['Program_encoded'] = le_program.fit_transform(df['Program'])
label_encoders['Program'] = le_program

le_status = LabelEncoder()
df['Status_encoded'] = le_status.fit_transform(df['Status'])
label_encoders['Status'] = le_status


def score_to_numeric(score):
    if pd.isna(score):
        return np.nan
    score_mapping = {
        'A': 4, 'A-': 3.7, 'B': 3, 'B-': 2.7, 'C': 2, 'C-': 1.7, 'D': 1, 'D-': 0.7, 'F': 0
    }
    scores = [score_mapping.get(s, np.nan) for s in score.split() if s in score_mapping]
    return np.nanmean(scores) if scores else np.nan

df['Score_processed'] = df['Score'].apply(score_to_numeric)

df['Score_processed'].fillna(df['Score_processed'].median(), inplace=True)
df_cleaned = df.drop(columns=['Program', 'Status', 'Score'])

df_cleaned.to_csv('C:\\Users\\HP\\Desktop\\walmart\\datasets\\updatedcdp.csv', index=False)


```
